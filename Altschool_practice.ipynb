```python
def roll_dice():
    # Generate random numbers for two dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    return die1, die2

def main():
    print("Welcome to the Dice Rolling Game!")
    while True:
        # Roll the dice
        die1, die2 = roll_dice()
        print(f"\nYou rolled: Die #1 = {die1}, Die #2 = {die2}")

        # Ask user if they want to continue
        user_input = input("Would you like to roll again? (yes/no): ").strip().lower()
        while user_input not in ['yes', 'no']:
            user_input = input("Please enter 'yes' or 'no': ").strip().lower()
        if user_input == 'no':
            print("Thank you for playing! Goodbye.")
            break

if __name__ == "__main__":
    main()
```

```python
import random

def roll_dice():
    # Simulate rolling two dice
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    print(f"\nYou rolled:\n - Dice #1: {die1}\n - Dice #2: {die2}")
    print(f"Total: {die1 + die2}")

def main():
    print(" Welcome to the Dice Rolling Game! ")

    while True:
        input("\nPress Enter to roll the dice... ")
        roll_dice()

        # Ask if the user wants to play again
        play_again = input("\nWould you like to roll again? (yes/no): ").strip().lower()
        if play_again not in ['yes', 'y']:
            print("\nThanks for playing! Goodbye!")
            break

if __name__ == "__main__":
    main()
```

```python
# Initialize an empty list to store numbers
numbers = []

while True:
    user_input = input("Pick a lucky number (or enter 'e' to exit): ")

    if user_input.lower() == 'e':
        break

    try:
        num = float(user_input)  # Use float to allow decimals too
        numbers.append(num)
    except ValueError:
        print("Invalid input! Please enter a number or 'e' to exit.")

# Calculate and print average if numbers were entered
if numbers:
    average = sum(numbers) / len(numbers)
    print(f"\nAverage of entered numbers: {average}")
else:
    print("\nNo numbers were entered.")
```
